Matrix Operations:


• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:
1. Prove A.A−1 = I.
2. Prove AB ̸= BA.
3. Prove (AB)T = B^TA^T
.

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

````python
import numpy as np
#1
A = np.array([[3,4], [7,8]])
B = np.array([[5,3],[2,1]])

A_inverse = np.linalg.inv(A)
#To prove A.A−1 = I:
result = np.matmul(A,A_inverse)
print("Multiplication of A and inverse of A: \n",result, "\n")


#identity matrix
identity_matrix = np.eye(2)
print("Identity matrix: \n", identity_matrix, "\n")
print("A.A−1 = I: ",np.allclose(result, identity_matrix), "\n")


#2
AB = np.dot(A,B)
print("A.B : \n", AB, "\n")
BA = np.dot(B,A)
print("B.A : \n", BA, "\n")
print("AB == BA ? : ", np.allclose(AB,BA), "\n")

#3
transpose_of_AB = np.transpose(AB)
print("Transpose of AB: \n",transpose_of_AB, "\n")
transpose_of_B = np.transpose(B)
transpose_of_A = np.transpose(A)
A_transpose_b_transpose = np.dot(transpose_of_B, transpose_of_A)
print("Atranspose.Btranspose : \n", A_transpose_b_transpose, "\n")
print(" (AB)T == B^TA^T ?: ", np.allclose(transpose_of_AB, A_transpose_b_transpose), "\n")


#4
# 2x − 3y + z = −1
# x − y + 2z = −3
# x + y − z = 9

a1 = np.array([[2, -3, 1],
              [1, -1, 2],
              [1,  1, -1]])

b1 = np.array([[-1],
              [-3],
              [9]])

a_inverse = np.linalg.inv(a1)
print("Inverse of A: ", a_inverse, "\n")
x = np.dot(a_inverse, b1)
print("Solution X:\n", x)
````
